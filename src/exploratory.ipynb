```python
import re

import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta

def convert_to_standard_timestamp(input_date,output_date, field_name, df_name):
    return np.where(df_name[field_name]==input_date,output_date,df_name[field_name])

pat = re.compile('plad.*$')

```

```python
print(data.shape)
data = data.loc[data.Curdate.notnull() & data['Phone Numbers'].notnull()]
data = data.loc[data['Placement AD']!='x']
data = data.loc[data['Send?']==True]

print(data.shape)

data.head()

```

```python
data['Curdate'] = convert_to_standard_timestamp('18th Sep 09:11','18/09/2019 09:11:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('22nd Sep 10:27','22/09/2019 10:27:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('22nd Sep 10:27','22/09/2019 10:27:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('2019-09-16T16:14:00.000+01:00','22/09/2019 10:27:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('2019-09-16T17:36:01.000+01:00','22/09/2019 10:27:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('16th Sep 18:39','16/09/2019 18:39:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('17th Sep 09:08','17/09/2019 09:08:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('17th Sep 10:02','17/09/2019 10:02:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('17th Sep 10:16','17/09/2019 10:16:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('17th Sep 13:12','17/09/2019 13:12:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('17th Sep 13:41','17/09/2019 13:41:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('17th Sep 15:17','17/09/2019 15:17:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('17th Sep 16:53','17/09/2019 16:53:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('18th Sep 10:42','18/09/2019 10:42:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('18th Sep 10:58','18/09/2019 10:58:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('18th Sep 12:59','18/09/2019 12:59:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('18th Sep 14:56','18/09/2019 14:56:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('18th Sep 16:10','18/09/2019 16:10:00','Curdate',data)
data['Curdate'] = convert_to_standard_timestamp('18th Sep 16:46','18/09/2019 16:46:00','Curdate',data)

data['Curdate'] = pd.to_datetime(data['Curdate'],format='%d/%m/%Y %H:%M:%S')
```

```python
print(data.shape)
data.drop_duplicates(inplace = True)
print(data.shape)
```

```python
carer_df = pd.DataFrame()
data.reset_index(inplace=True)

print(datetime.now())

for i,r in data.iterrows():

    nr = {}
    print(f'{i}/{len(data)}')
    for p in r["Phone Numbers"].replace('\n',',').split(','):
        carer_id_row = ids.loc[(ids.raw_phone_number==p)|(ids.formatted_phone_number==p)]['professional_id'].values
        if len(carer_id_row)==1:
            nr['carer_id'] = carer_id_row[0]

        nr['raw_phone_number'] = p
        nr['sent_at'] = r['Curdate']
        nr['placement_ad_id'] = pat.findall(r['Placement AD'])[0]
        carer_df = carer_df.append(nr, ignore_index=True)
print(datetime.now())
```
