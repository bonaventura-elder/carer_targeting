```python
carer_df = pd.DataFrame()
data.reset_index(inplace=True)

print(datetime.now())

for i,r in data.iterrows():

    nr = {}
    print(f'{i}/{len(data)}')
    for p in r["Phone Numbers"].replace('\n',',').split(','):
        carer_id_row = ids.loc[(ids.raw_phone_number==p)|(ids.formatted_phone_number==p)]['professional_id'].values
        if len(carer_id_row)==1:
            nr['carer_id'] = carer_id_row[0]

        nr['raw_phone_number'] = p
        nr['sent_at'] = r['Curdate']
        nr['placement_ad_id'] = pat.findall(r['Placement AD'])[0]
        carer_df = carer_df.append(nr, ignore_index=True)
print(datetime.now())
```
